#  Preprocessing Data - Renaming columns and Using Lower Case

```python
covid_ind_df = covid_ind_df.rename(columns = {'State/UnionTerritory':'state'})
covid_ind_df.columns = map(str.lower, covid_ind_df.columns)
```

```python
covid_ind_df.columns
```

```python
sorted_state_df = covid_ind_df.sort_values('confirmed', ascending = False)
sorted_state_df.head(10)
```

```python
import plotly.graph_objects as go

def state_covid_plot(state):
    labels = ['confirmed', 'deaths','cured']
    colors = ['blue','red','green']
    
    line_size = [4,5,6]
    
    fig = go.Figure()
    
    
    state_df = covid_ind_df.loc[covid_ind_df['state'] == state]
    x_confirm_data = state_df['date']
    y_confirm_data = state_df['confirmed']
    y_death_data = state_df['deaths']
    y_cured_data = state_df['cured']
    
    x_data = np.array(list(x_confirm_data))
    deaths =  np.array(list(y_death_data))
    confirm = np.array(list(y_confirm_data))
    cured = np.array(list(y_cured_data))
    
    df_list = [confirm, deaths,cured]
    
    for i, data in enumerate(df_list):
        fig.add_trace( go.Scatter (x=x_data, y=data, mode='lines+markers', 
                                   name = labels[i], line = dict(color = colors[i], width = line_size[i]), 
                                  connectgaps = True, text = "Total" + str(labels[i]) + str(data[-1]) 
                                  ))
        fig.update_layout(title='COVID Impact In : '+state)
    fig.show()   
```

```python
state_list = list(covid_ind_df['state'].unique())
```

```python
 interact(state_covid_plot, state=state_list);
```
